# Session 9 Prompt
- Create a Jupyter notebook, import matplotlib.
- Use numpy to pull 1000 random numbers distributed uniformly between [0,1].
- Histogram the random numbers into 100 bins
- and plot the histogram.
- Label your axes and save the figure as a PDF.

```python
#import libraries as alias

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.uniform(0,1,1000)

#create the histogram bins; 100
width = 0.01
bins = np.arange(0,1,width)

#plot the histogram
plt.hist(x,bins=bins, alpha=0.35, edgecolor='k')

#label axes
plt.xlabel('x')
plt.ylabel('N per bin')
plt.title('Session 9')

#save the figure as a pdf
plt.savefig('astr19_session9_histogram.pdf', bbox_inches = 'tight', pad_inches = 0, dpi = 600)
plt.show()
```
